```python
epsilon = np.random.normal(loc=0.0, scale=4.0, size=1000)
X = []
for i in range(1000):
    X.append(random.uniform(21, 65))
X = np.array(X, dtype="float32").reshape(-1,1)
y = 1.5*X + epsilon.reshape(-1,1)

scaler1 = StandardScaler()
X = scaler1.fit_transform(X)
scaler2 = StandardScaler()
y = scaler2.fit_transform(y)

class MLP(nn.Module):
  '''
    Multilayer Perceptron.
  '''
  def __init__(self):
    super().__init__()
    self.layers = nn.Sequential(
      #nn.Sigmoid(),
      nn.Linear(1, 1),
      nn.LeakyReLU(1), 
      nn.Linear(1,1)
      #nn.ReLU()
    )


  def forward(self, x):
    '''Forward pass'''
    return self.layers(x)
  
  
if __name__ == '__main__':
  
  # Set fixed random number seed
  #torch.manual_seed(42)
  
  # Prepare CIFAR-10 dataset
  dataset = (np.vstack((X,y))).T
  trainloader = torch.utils.data.DataLoader(dataset, batch_size=1000, shuffle=True, num_workers=1)
  
  # Initialize the MLP
  mlp = MLP()
  
  # Define the loss function and optimizer
  loss_function = nn.MSELoss()
  optimizer = torch.optim.SGD(mlp.parameters(), lr=0.1, momentum=0.30)
  
  LR_scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=0.1, patience=25, threshold=5, threshold_mode='abs', eps=1e-16, verbose=True)
  
  # Run the training loop
  for epoch in range(0, 30): # 5 epochs at maximum
    
    # Print epoch
    print(f'Starting epoch {epoch+1}')
    
    # Set current loss value
    current_loss = 0.0
    
    # Iterate over the DataLoader for training data
    for i, data in enumerate(trainloader, 0):
      #print(data)
      # Get inputs
      #inputs = torch.index_select(data, 1, torch.tensor([0])).type(torch.FloatTensor)
      #targets = torch.index_select(data, 1, torch.tensor([1])).type(torch.FloatTensor)
      inputs = torch.tensor(X).type(torch.FloatTensor)
      targets = torch.tensor(y).type(torch.FloatTensor)
      #print(targets)  
    #print(targets)
      #print(targets.shape)
      # Zero the gradients
      optimizer.zero_grad()
      
      # Perform forward pass
      outputs = mlp(inputs)
      
      # Compute loss
      loss = loss_function(outputs, targets)
      
      # Perform backward pass
      loss.backward()
      
      # Perform optimization
      optimizer.step()
      
      # Print statistics
      current_loss = loss.item()
      #LR_scheduler.step(current_loss)
    
      print(f'Current Loss {current_loss}')
  # Process is complete.
  print('Training process has finished.')
```
